```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
```

```python
df = pd.read_excel(r'f001_1.xlsx', header=6)
df = df.iloc[2:,1:]
fn = df[['項目名', '1_居住世帯あり']]
fn = fn.set_index(keys='項目名')
fn['1_居住世帯あり'] = pd.to_numeric(fn['1_居住世帯あり'], errors='coerce')
fg, ax = plt.subplots(figsize=(22,5))
ax1 = ax.twinx()
sns.heatmap(data=fn.T,
            ax=ax,
            linewidths=0.3,
            cmap='YlGnBu',
            linecolor='lightgray', alpha=0.7)

ax1.plot(fn, color='tomato', alpha=0.7)
ax.tick_params(axis='x', rotation=90)
ax.set_xlabel('地域区分')
ax.grid(True)
```
